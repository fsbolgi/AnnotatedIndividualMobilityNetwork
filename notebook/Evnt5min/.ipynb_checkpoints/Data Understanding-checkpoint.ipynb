**Count the number of distinct enginestatus values**

```python
cur.execute("SELECT enginestatus, COUNT(*) as count FROM tak.vodafone_zel1_evnt_5min GROUP BY enginestatus ORDER BY count DESC")
```

```python
ar = []

for f in cur:
    ar.append(list(f))

print(ar)
```

```python
fig = plt.figure(figsize=(7, 7)) 
fig_dims = (1, 1)

ax = plt.subplot2grid(fig_dims, (0, 0))
labels = 'enginestatus = 3', 'enginestatus = 2'
sizes = [109988637, 62343353]
explode = (0.01, 0.05)
c = ["#97c170", "#dde37a"]
inside, texts, ltexts = ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=50, colors=c)
for i in range(len(texts)):
    texts[i].set_fontsize(13)
    ltexts[i].set_fontsize(13)
ax.axis('equal') 

plt.savefig('../../../thesis/images/Evnt5_engine.png', format='png', bbox_inches='tight')
```
