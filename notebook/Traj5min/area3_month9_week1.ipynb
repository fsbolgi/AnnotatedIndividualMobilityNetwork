```python
# folium allows to visualize geolocalized data with the jupyter notebook
import folium
from folium.plugins import MarkerCluster
# pandas is needed to read the csv file and to perform some basic operations on dataframes
import pandas as pd
# matplotlib is used to draw graphs
import matplotlib.pyplot as plt
# module for different stats
from statistics import mean 
```

```python
stop = "5"
```

```python
df = pd.read_csv('../../../datasets/in/Traj'+stop+'min/area3_month9_week1_stops.csv') 
```

```python
print("The dataset has", df.shape[0], "rows and", df.shape[1], "columns.")
```

```python
vehicle_list = df.vehicle.unique()
```

```python
v = vehicle_list[0]
df_v = df[df["vehicle"] == v]
print("The dataset has", df_v.shape[0], "rows and", df_v.shape[1], "columns.")
```

```python
df_v.head()
```

```python
colors = ['gray', 'orange', 'green', 'lightred', 'darkgreen', 'darkblue', 'darkred', 'pink', 
          'lightgrayblack', 'lightblue', 'cadetblue', 'darkpurple', 'lightgreen', 'purple', 'blue', 'red', 'beige']

m = folium.Map(location=[40, 20], zoom_start=8)

points_strings = list(df_v["start_point"]) + list(df_v["end_point"])

points = []

for p in points_strings:
    p_0 = (p.split(','))[0][1:]
    p_1 = (p.split(','))[1][1:-1]
    points.append([p_0, p_1])
    
mc = MarkerCluster()
for i in range(len(df_v)):
    marker = folium.Marker(location=points[i], popup=i, icon=folium.Icon(color=colors[i%17],icon='play'))
    mc.add_child(marker)
    
    marker = folium.Marker(location=points[i], popup=i, icon=folium.Icon(color=colors[i%17],icon='stop'))
    mc.add_child(marker)

m.add_child(mc)
m
```
