### This time we use the csv with the coordinates to try to plot them

```python
df_coord = pd.read_csv("../../../datasets/out/Traj5min/vehicle_3270_79350_coord.csv") 
```

```python
def swapXY(traj):
    new_traj = []
    for i in traj:
        new_traj.append([i[1], i[0]])
    return new_traj
```

```python
coords = []

for el in  df_coord["trajcoord"]:
    y = json.loads(el)
    c = swapXY(y["coordinates"])
    coords.append(c)
```

```python
df_coord["coord"] = coords
```

```python
df_coord.head(5)
```

The majority of trajectories are not relevant. We can consider all trajectories beaneath a certain threashold just as GPS errors. In particular we consider just trajectories where:
- _duration_ is at least 120 sec;
- _length_ is at least 0.2 km;

```python
#remove trajectories of less than 2 minutes
df_long = df_coord[df_coord["duration"] > 120]
#remove trajectories of less than 200 meters
df_long = df_long[df_long["length"] > 0.2]
```

```python
print ("Of the total", df.shape[0], "just", df_long.shape[0], "are trajectories are significant")
```

```python
df_long = df_long.drop(columns=['trajcoord'])
```

```python
df_long.head(5)
```

**The problem is that we can't draw all the trajectories together, it's a computation too heavy**

We can split it for periods of time

```python
print("the first trajectory starts at", min(df_long["start_time"]))
print("the ast trajectory finish at", max(df_long["start_time"]))
```

So we are covering the period from the first of July 2017 to the 28 of June 2018. Let's draw a map for each month.

```python
df_long['start_month'] = pd.DatetimeIndex(df_long['start_time']).month
```

**JULY 2017**

```python
df_july = df_long[df_long["start_month"] == 7]
```

```python
print ("in july this vehicle has", len(df_july), "significant trajectories")
```

```python
coords = df_july["coord"].tolist()
```

```python
colors = ['gray', 'orange', 'green', 'lightred', 'darkgreen', 'darkblue', 'darkred', 'pink', 
          'lightgrayblack', 'lightblue', 'cadetblue', 'darkpurple', 'lightgreen', 'purple', 'blue', 'red', 'beige']

m = folium.Map(location=[38, 22], zoom_start=8.4)
for i in range(202):
    my_PolyLine=folium.PolyLine(locations=coords[i], weight=5, color=colors[i%17])
    m.add_child(my_PolyLine)
    
mc = MarkerCluster()
for i in range(202):
    marker = folium.Marker(location=coords[i][0], popup=i, icon=folium.Icon(color=colors[i%17],icon='play'))
    mc.add_child(marker)
    
    marker = folium.Marker(location=coords[i][-1], popup=i, icon=folium.Icon(color=colors[i%17],icon='stop'))
    mc.add_child(marker)

m.add_child(mc)
m
```

**NOVEMBER 2017**

```python
df_novem = df_long[df_long["start_month"] == 11]
```

```python
print ("in november this vehicle has", len(df_novem), "significant trajectories")
```
